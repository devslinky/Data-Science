## Q2(a)

```python
import re
```

```python
sample_text = """
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases are not special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way to do it.
Although that way may not be obvious at first unless you are Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it is a bad idea.
If the implementation is easy to explain, it may be a good idea.
Namespaces are one honking great idea -- let us do more of those!

""" # The Zen of Python by Tim Peters
```

```python
def wordCount(text):

    list_of_words = re.split("\.+", re.sub('[^a-zA-Z0-9]', '.', text).strip('.'))
    ## your code starts here ##

    # initialize empty dictionary
    word_dict = {}

    # make all words in list_of_words undercase
    list_of_words = list(map(lambda x: x.lower(), list_of_words))

    # generates a list of "unique" words using set()
    unique_words = list(set(list_of_words))

    counter = 0

    # checks number of occurences of a unique word and tallys them
    # adds the tally as the value and the word, the key
    for unique_word in unique_words:
      for word in list_of_words:
        if unique_word == word:
          counter += 1
      word_dict[unique_word] = counter

    return word_dict
    ## end ##
```

```python
# example usage
wordCount(sample_text)
```

```python
def myFuncDictFilter(input_dict, n):
    ## your code starts here ##
    return {key:value for (key,value) in input_dict.items() if value > n }
    ## end ##
```

```python
# example usage
myFuncDictFilter(wordCount(sample_text), 2)
```
